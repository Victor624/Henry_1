```python
import pandas as pd
import ast
import numpy as np
import os
```

```python
df = pd.read_csv("frac_movie.csv")
```

# Desanidado de columnas 

## Comenzamos con la columna Belongs

```python
data=df["belongs_to_collection"]
data=data.str.replace("{","")
data=data.str.replace("}","")
data=data.str.replace("'id':","")
data=data.str.replace("'name':","")
data=data.str.replace("'poster_path':","")
data=data.str.replace("'backdrop_path':","")
data=pd.DataFrame(data)
data.fillna('0 , 0 , 0 , 0', inplace=True)
data=data["belongs_to_collection"]
```

```python

def Separar(lista):
    dat=lista.split(",")
    dat=pd.DataFrame(dat)
    dat=dat.T
   
    return dat
```

```python
belong=pd.DataFrame()
belong= pd.concat(list(map(lambda i: Separar(i), data)))
belong
```

```python

belong['id_belong']=belong[0]
belong['name_belong']=belong[1]
belong=belong[['id_belong','name_belong']]
```

## Desanidando Genero


```python
genres= df['genres']
```

```python
def funcion_t(x):
    a=pd.DataFrame({"a":[x]})
    return a
    
```

```python
def Desa_Gen(x):
    
    genres=funcion_t(x)
    genres=pd.DataFrame(genres)
    a=genres.columns[0]
    genres=genres.rename(columns={ a:'genres'})

    genres['genres'] = genres['genres'].apply(ast.literal_eval) 
    df_exploded = genres.explode('genres', ignore_index=True) 
    df_exploded['name_genres'] = df_exploded['genres'].apply(lambda x: x.get('name', '') if isinstance(x, dict) else '')
    df_exploded['id_genres'] = df_exploded['genres'].apply(lambda x: x.get('id', '') if isinstance(x, dict) else '')
    df_exploded=df_exploded.drop(columns="genres")
    data=df_exploded.T

    dato = data.apply(lambda x: ' '.join(x.astype(str)), axis=1)
    dato=pd.DataFrame(dato)
    dato=dato.T[["name_genres","id_genres"]]
    return dato
```

```python
genr=pd.DataFrame()
genr=pd.concat(list(map(lambda i: Desa_Gen(i), genres)))
genr
```

## Desanidando Spoken

```python
data=df["spoken_languages"]
data.fillna('0 , 0 , 0 , 0', inplace=True)
```

```python
def Desa_Spok(x):
    
    Spok=funcion_t(x)
    Spok=pd.DataFrame(Spok)
    a=Spok.columns[0]
    Spok=Spok.rename(columns={ a:'Spok'})

    Spok['Spok'] = Spok['Spok'].apply(ast.literal_eval) 
    df_exploded = Spok.explode('Spok', ignore_index=True) 
    df_exploded['iso_639_1_Spoken'] = df_exploded['Spok'].apply(lambda x: x.get('iso_639_1', '') if isinstance(x, dict) else '')
    df_exploded['name_Spoken'] = df_exploded['Spok'].apply(lambda x: x.get('name', '') if isinstance(x, dict) else '')
    df_exploded=df_exploded.drop(columns="Spok")
    data=df_exploded.T

    dato = data.apply(lambda x: ' '.join(x.astype(str)), axis=1)
    dato=pd.DataFrame(dato)
    dato=dato.T[["name_Spoken","iso_639_1_Spoken"]]
    return dato
```

```python
spoken=pd.DataFrame()
spoken=pd.concat(list(map(lambda i: Desa_Spok(i), data)))
spoken
```

## Desanidando Production Companies

```python
data=df["production_companies"]
data.fillna('0 , 0 ', inplace=True)
```

```python
def Desa_Prod(x):
    
    Prod=funcion_t(x)
    Prod=pd.DataFrame(Prod)
    a=Prod.columns[0]
    Prod=Prod.rename(columns={ a:'Prod'})

    Prod['Prod'] = Prod['Prod'].apply(ast.literal_eval) 
    df_exploded = Prod.explode('Prod', ignore_index=True) 
    df_exploded['id_prod'] = df_exploded['Prod'].apply(lambda x: x.get('id', '') if isinstance(x, dict) else '')
    df_exploded['name_prod'] = df_exploded['Prod'].apply(lambda x: x.get('name', '') if isinstance(x, dict) else '')
    df_exploded=df_exploded.drop(columns="Prod")
    data=df_exploded.T

    dato = data.apply(lambda x: ' '.join(x.astype(str)), axis=1)
    dato=pd.DataFrame(dato)
    dato=dato.T[['id_prod','name_prod']]
    return dato
```

```python
Produ=pd.DataFrame()
Produ=pd.concat(list(map(lambda i: Desa_Prod(i), data)))
Produ
```

## Desanidando Production Countries

```python
data=df["production_countries"]
data=pd.DataFrame(data)
data["production_countries"] = data["production_countries"].mask( data["production_countries"].str.contains(r'^\[(.*?)\]$') == False, '[]')
data.fillna("[{'iso_3166_1': 'null', 'name': 'null'}]", inplace=True)
data=data["production_countries"]
```

```python
def Desa_Count(x):
    
    Count=funcion_t(x)
    Count=pd.DataFrame(Count)
    a=Count.columns[0]
    Count=Count.rename(columns={ a:'Count'})

    Count['Count'] = Count['Count'].apply(ast.literal_eval) 
    df_exploded = Count.explode('Count', ignore_index=True) 
    df_exploded['iso_3166_1_Count'] = df_exploded['Count'].apply(lambda x: x.get('iso_3166_1', '') if isinstance(x, dict) else '')
    df_exploded['name_Count'] = df_exploded['Count'].apply(lambda x: x.get('name', '') if isinstance(x, dict) else '')
    df_exploded=df_exploded.drop(columns="Count")
    data=df_exploded.T

    dato = data.apply(lambda x: ' '.join(x.astype(str)), axis=1)
    dato=pd.DataFrame(dato)
    dato=dato.T[['iso_3166_1_Count','name_Count']]
    return dato
```

```python
Count=pd.DataFrame()
Count=pd.concat(list(map(lambda i: Desa_Count(i),data )))
Count
```

## Concatenando los desanidados de Movies para continuar trabajando

```python
Desa_Movie=pd.concat([belong,genr,spoken,Produ,Count],axis=1)
Desa_Movie.head(2)
```
